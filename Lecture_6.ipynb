Write a program that asks the user for a positive integer n. Calculate and print the product of the
integers 1 through n. 

```python
n = int(input("Enter a positive integer n"))
product = 1
i = 1
while i <= n:
    product *= i
    i+=1
print(product)
```

Write a program that asks the user for a positive integer n. Calculate and print the sum of the
odd integers in the range 1 through n. 

```python
n = int(input("Enter a positive integer n"))
sum = 0
i = 1
while i <= n:
    if(i%2==1):
        sum += i
    i+=1
print(sum)
```

Write a program to request and validate a password from your user. A valid password
will be any two digit integer, both digits of which are even.
Give the user three chances to enter a correct password.

• At each incorrect attempt, print “Invalid password. Try again”

• If the password entered is correct print “Correct! You may access the system.” Exit the
program.

• If the password entered is incorrect print “Too many invalid attempts. Please try again
later.”

```python
password = int(input("Please enter a valid password"))

tries = 1
while tries < 3:
    if(password > 10 and password < 99):
        if((password%10)%2==0 and (password//10)%2==0):
            print("Correct! You may access the system")
            tries = 4
        else:
            print("Invalid password. Try again")
            password = int(input("Invalid password. Try again: "))
            tries += 1
    else:
        print("Invalid password. Try again")
        password = int(input("Invalid password. Try again: "))
        tries += 1
if tries == 3:
    print("Too many invalid attemps. Please try again later.")
```

Write a program that asks the user for a positive integer n where the right-most digit is not a
zero. Construct and output the integer whose digits are the reverse of those in n.

For example, if n has the value 123, then you need to construct the integer 321. Note you are not
just printing the digits of n in reverse order; you are actually constructing the new integer. 

```python
n = int(input("Enter a positive interger n where the right-most digit is not a zero"))
if(n%10==0):
    print("Wrong! It should not end in 0!")
else:
    num = 0
    while n != 0:
        num *= 10
        num += n %10
        n //= 10
    print(num)
```

Write a program that inputs a positive integer n and determines if n is prime or composite (i.e.
not a prime). 

```python
n = int(input("Enter a number"))
is_prime = True
i = 2
while i != n and is_prime:
    if(n%i==0):
        is_prime = False
    i += 1
print(f"{n} is prime!" if is_prime else f"{n} is not prime!")
```

For Loops

Syntax:

for <some variable> in range(…) :
 
 one or more statements (called a code block)

range(…) has the following syntax: 
range([start value,] end value [,step]) 
 

```python
for i in range(5):
    print(i)
```

This will assign i the values 0 -> 5 not including 5!

```python
for i in range(1, 4):
    print(i)
```

will print numbers 1 -> 3 not including 4!

```python
for i in range(1, 10, 2):
    print(i)
```

This will print numbers 1 -> 9 skipping every 2 numbers not including 10

```python
for i in range(5, 1, -1):
    print(i)
```

this will print numbers 2 -> 5 not including 1 backwards!

Generate and print all numbers between 1 and 1000 such that the sum of the digits equals 20. 


```python
for i in range(1, 101):
    sum = 0
    num = i

    while num > 0:
        sum += num%10
        num //= 10

    if(sum==20):
        print(i)
```

Print all prime numbers between 1 and 100

```python
for i in range(1, 101):
    num = i
    is_prime = True
    for j in range(2, num):
        if(num%j==0):
            is_prime = False
    if is_prime:
        print(num)
```

Write a program, using for loops, to generate the following output. 



```python
for i in range(1, 10):
    for j in range(1, 10):
        if j >= i:
            print(j, end="")
        else:
            print(" ", end="")
    for j in range(8, i-1, -1):
        print(j, end="")
    print()
```
